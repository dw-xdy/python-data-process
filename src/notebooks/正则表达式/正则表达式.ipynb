## 正则表达式中的**元字符**

对于正则表达式来说, 最重要的就是这几个元字符.
### 1  **.**  通配符, 匹配除了换行之外的的任意的**单个**字符.

````python
import re

s = "apple ape agree age amaze animate advertise a\ne"

ans = re.findall("a.e", s)  # 需要注意的是: 并不匹配换行符, 所以 a\ne 并不符合查找条件.
result = re.findall("a..e", s)

print(ans)
print(result)
# 从这里也能看出来正则表达式匹配的结果是: list类型, 其中全部都是字符串.
print(type(result))
for res in result:
    print(type(res))
````

### 2.  **[]** 字符集, 匹配集合中所含有的任意一字符 (并不需要加 , 分割). 
### 3.  **-** 范围限定符, 在[](字符集)内部, 并且不在开头结尾的时. 定义范围.  
### 4.  **^** 取反符, 在[](字符集)内部, 并且一定是要在字符集的开头位置, 表示对当前的字符集进行取反匹配.  
(这三个经常一起使用, 所以放在一起学习)

````python
s = "apple ape agree age amaze fanimate advertise a\ne a&e a@e a6e a9e"

result = re.findall("a.e", s)  # 将result拿下来做一个对比, 若是使用 . 那就有点太宽泛了.
ans = re.findall("a[apc]e", s)  # 这里的 a[apc]e 表示 a 和 e 中间只能是 a p c 其中之一.

# 这里的 - (范围限定符) 限定了范围: a e 中间的符号只能是: a ~ z 或者 A ~ Z 中的一个, 相当于只在这52个字符中选择.
res = re.findall("a[a-zA-Z]e", s)

# 这里的 ^ 取反, 即 匹配任意不在字符集中的内容, ^ 必须放在字符集的开始位置.
res_out_character = re.findall("a[^a-zA-Z]e", s)
res_out_num = re.findall("a[^0-9]e", s)
res_out_character_and_num = re.findall("a[^0-9a-zA-Z]e", s)

print(result)
print(ans)
print(res)
print(res_out_character)
print(res_out_num)
print(res_out_character_and_num)
````

### 重复选择符: {} * + ?  (最核心的!!!)
### 5. {} 数量范围贪婪符 指定左边原子的可以重复的数量范围 (这里会涉及到一个贪婪匹配, 非贪婪匹配的问题, 需要非常注意!!!)
    四种写法: {n}, {n,} {,m} {n,m} 四种写法, 
### 6. * 指定左边原子出现 0 次或多次, 等同于: {0,}
### 7. + 指定左边原子出现 1 次或多次, 等同于: {1,}
### 8. ? 指定左边原子出现 0 次或 1 次, 等同于: {0,1} (这个 ? 也可以用来取消贪婪匹配, 但是只是使用的位置不同.)

````python
# {} 数量选择符的代码和理解.

s = "aeeee apple acre ape agree age amaze animate advertise a\ne a&e ade a6e a9e"
# ret = re.finall("a...e", s)
ret = re.findall("a.{3}e", s)  # 这段代码的效果和上面代码的效果完全一样.

"""
res最后的结果是: ['aeeee', 'apple', 'acre', 'ape', 'agree', 'age', 'amaze', 'ate', 'adve', 'alne', 'a&e', 'ade', 'a6e', 'a9e']
注意这里遇到的问题, aee, aeee, aeeee, 那么对应的, 应该匹配哪一个呢? 从结果我们可以知道, 最后的结果是aeeee. 没有匹配前面比较短的.
所以这里默认的 贪婪匹配  按照最大匹配进行匹配 
贪婪匹配解释: 比如这个: aeeee, 既然可以匹配上 a...e, 那么之后我就直接跳过了, 不进行匹配aee, aeee了, 不管了, 直接匹配下一个去.
贪婪匹配解释: 按照最大的匹配数进行优先匹配.
取消贪婪匹配: 按照最小的匹配数进行优先匹配. (操作: 在重复符后面跟着一个 ? )
还有需要注意的是: 正则表达式的匹配是: 匹配完成了一个之后, 立刻将其抛出, 然后继续往后走, 匹配下一个. 
比如说非贪婪匹配, 匹配到了aee之后, 同样并不会匹配 aeee, aeeee.
"""

# res = re.findall("a.{1,3}e", s) # 需要注意的是: 这里的范围表示 1 ~ 3. 并不是 1 或 3 (并且 , 之后是没有空格的)
res = re.findall(
    "a.{1,3}?e", s
)  # 需要注意的是: 这里的范围表示 1 ~ 3. 并不是 1 或 3 (并且 , 之后是没有空格的)
# 这里的 {1,} 表示匹配无穷大, 相当于后面的数字是 无穷大.
ans = re.findall("a.{1,}e", s)  # 匹配出了两个字符串, 因为 . 无法匹配 \n
ans1 = re.findall(
    "a.{1,}e", s, re.S
)  # 若是有匹配 \n 的需求, 那么就在 第三个参数的位置加上: re.S
ans2 = re.findall(
    "a[a-z]{1,3}e", s
)  # 不进行匹配特殊字符, 这里其实分开拆解比较好理解: 先有一个字符集, 只能匹配a-z, 然后这个字符集可以出现1~3次.
# 相当于是: 匹配中间是: a ~ e 的字符串, 并且这些字符串可以进行匹配 1 ~ 3 次, 并且按照最大匹配长度进行优先匹配.

ss = "ae aeeee apple acre ape agree age amaze animate advertise a\ne a&e ade a6e a9e"
# 这里需要非常注意: 因为有 0 这个选项, 但是此时是贪婪匹配, 优先匹配最长的, 所以匹配到的结果是: "ae ae", 并不是 ae.
# 要按照规则来, 这是字符串, 空格都是字符串的一个字符, 所以要充分理解, 而不是理解为: 空格是用来隔开字符的, 并不是, 空格也是字符.
ans3 = re.findall("a.{0,3}e", ss)

print(ret)
print(res)
print(ans)
print(ans1)
print(ans2)
print(ans3)
````

````python
# * 符号的代码与解释 ( * 只是 {0,} 的简写, 两个符号的效果完全一样.)
#  .*? 这一组格式我们是经常用到的. 所以一定要完全理解其意义.
# 还有要补充的一点是: []*? 这个组合的语法和 .*? 是类似的, 只是 [] 字符集缩小了范围而已.

s = "aeeee apple acre ape agree age amaze animate advertise a\ne a&e ade a6e a9e"
# 贪婪匹配!!!
ret = re.findall("a.{0,}e", s)
# 我们在实际工作中, 一般使用的是: 非贪婪匹配与 {0,}搭配, 因为一般是找到 ae 距离最近 (a e中间夹着的内容)
ret1 = re.findall("a.{0,}?e", s)

# 贪婪匹配!!!
ans = re.findall("a.*e", s)
# 我们在实际工作中, 一般使用的是: 非贪婪匹配与 {0,}搭配, 因为一般是找到 ae 距离最近 (a e中间夹着的内容)
ans1 = re.findall("a.*?e", s)

# 一般来说, .*? 这个模式匹配一般是用来提取数据, 并进行数据清洗的, 具体可以看视频, 因为我现在还没有学习爬虫, 等我学习完了爬虫之后可能会有更深入的理解吧.

print(ret)
print(ret1)

print(ans)
print(ans1)
````

````python
# * 符号的代码与解释 ( * 只是 {1,} 的简写, 两个符号的效果完全一样.)

#         注意 要注意这个 ae apple 因为我们使用的是 .+? 所以会导致一个问题, 就是 ae 我们是没办法匹配的, 所以往后走, 最后的结果是: ae apple
s = "aeeee ae apple acre ape agree age amaze animate advertise a\ne a&e ade a6e a9e"

res = re.findall("a.+e", s)  # 贪婪匹配
ans = re.findall("a.+?e", s)  # 非贪婪匹配

print(res)
print(ans)
````

````python
# ? 符号的代码与解释 ( ? 只是 {0,1} 的简写, 两个符号的效果完全一样.)
# ( ? 这个符号容易和 是否取消贪婪匹配的 ? 混淆, 所以要注意!!!)

s = "https://www.baidu.com/,yuan,rain,http://www.jd.com/,https://www.taobao.com/"

ans = re.findall("https??://www.[a-z]*?.com", s)

ss = "yuan,rain,http://www.jd.com/,https://www.taobao.com/,https://www.baidu.com/"

ret = re.findall("https??://www.[a-z]*?.com", ss)

print(ans)
print(ret)
````

````python
# 贪婪模式匹配的例子

# 贪婪模式匹配的例子同样很常见, 应用也比较广泛, 但是还是非贪婪匹配应用更加广泛

s = "yuan 23 rain 188 alvin 8888 eric 9999"

# 这里使用了贪婪匹配, 若是使用非贪婪匹配最后的结果会导致所有的数字单独被提取出来.
ans = re.findall("[0-9]+", s)
ret = re.findall("[0-9]+?", s)

print(ans)
print(ret)
````

### 9. ^ 开始边界符 / 开始锚点符, 匹配一行的开头位置. (并不是取反, 只是相同的字符不同作用)
    ^ 必须在开头位置, 表示: 对应的开头位置前面不能有东西, 必须要以 ^ 之后的字符串开头, 若是没有, 那么直接匹配失败.
### 10. $ 结束边界符 / 结束锚点符, 匹配一行的结束位置.

    $ 必须在结尾位置, 表示: 对应的结尾位置后面不能有东西, 必须要以 $ 之前的字符串结尾, 若是没有, 那么直接匹配失败.

````python
path = "yuan/blog/2021/12"

#                         注意 : 这里{1,2}之后不能加 ? 因为, 若是按照非贪婪匹配的方式执行, 若是最后的数字是: 12, 那么就只能匹配到 1.
reg = "yuan/blog/[0-9]{4}/[0-9]{1,2}"

ans = re.findall(reg, path)

print(ans)
````

````python
path = "32987298yuan/blog/2021/12fdiosahoihdfs"

# 这里展示了 ^ $ 的作用,
reg = "^yuan/blog/[0-9]{4}/[0-9]{1,2}$"
# 若是没有 ^ $ 那么前面的跳过, 还可以接着匹配, 毕竟只是从中进行提取, 但是加了 ^ 和 $ 就不行了, 因为这两个符号会强制要求必须按照对应的字符串开头结尾.
reg1 = "yuan/blog/[0-9]{4}/[0-9]{1,2}"

ans = re.findall(reg, path)
ans1 = re.findall(reg1, path)

print(ans)
print(ans1)
````

### 11. \ 转义符 (但是需要注意: python中的转义符和正则表达式中的转义符是完全不同的两个概念!!!!!!)

#### **11.1 赋予某些符号特殊功能.**

##### **1. 字符类简写（常用）**


| 转义序列 | 正确写法 | 描述 | 等价形式 |
|---------|---------|------|---------|
| `\d` | `\d` | **数字字符** | `[0-9]` |
| `\D` | `\D` | **非数字字符** | `[^0-9]` |
| `\w` | `\w` | **单词字符**（字母、数字、下划线） | `[a-zA-Z0-9_]` |
| `\W` | `\W` | **非单词字符** | `[^a-zA-Z0-9_]` |
| `\s` | `\s` | **空白字符**（空格、制表符、换行等） | `[ \t\n\r\f\v]` |
| `\S` | `\S` | **非空白字符** | `[^ \t\n\r\f\v]` |

##### **2. 特殊字符转义**

| 转义序列 | 正确写法 | 描述 |
|---------|---------|------|
| `\n` | `\n` | **换行符** (newline) |
| `\t` | `\t` | **制表符** (tab) |
| `\r` | `\r` | **回车符** |
| `\f` | `\f` | **换页符** |
| `\v` | `\v` | **垂直制表符** |

##### **3. 边界匹配**

| 转义序列 | 正确写法 | 描述 |
|---------|---------|------|
| `\b` | `\b` | **单词边界**（单词与非单词之间） |
| `\B` | `\B` | **非单词边界** |
| `^` | `\^` | 字符串开始（需转义时才用`\^`） |
| `$` | `\$` | 字符串结束（需转义时才用`\$`） |

##### **4. 其他需要转义的元字符**
```
.  ^  $  *  +  ?  {  }  [  ]  \  |  (  )
```
这些字符在正则中有特殊含义，要匹配字面值需要加反斜杠：
```python
re.search(r'\.', 'a.b')    # 匹配点号
re.search(r'\$', '$100')   # 匹配美元符号
re.search(r'\\', 'a\\b')   # 匹配反斜杠
```




````python
# 首先我们还是先明确一下 python 中的转义字符和 正则表达式模块 re 中的不同
# 当我们执行这样一段代码的时候, 会导致报错, 因为这个 \d 会先被 python 解释器利用, 然后将其转入到 re 模块中进行解释,
# 但是因为python解释器可能会误处理, 导致警告, 因为python解释器不明白你要做什么.
s = "yuan 23 rain 188 alvin 8888 eric 9999"

# reg = "\d+"

# ans = re.findall(reg, s)

# print(ans)

# 所以此时我们有两种方式解决这个问题

# 第一种方式, 利用转义字符的特性, 将转义字符转义成普通字符.abs
reg = "\\d+"

ans = re.findall(reg, s)

print(ans)

# 第二种方式: 利用 r"", 这样做可以让python解释器完全按照原封不同的字符串进行处理, 消除转义字符的作用.abs

reg = r"\d+"

ans = re.findall(reg, s)

print(ans)
````

````python
# 这里重点了解一下 \b 的作用 (单词边界)

s = "The cat sat on the caterpillar."

ans = re.findall("cat", s)

# 需要注意的是: 如果我们这里只想要拿到单词, 而不是单词的其中一部分, 比如这里 caterpillar 单词中也有 cat, 我们也拿到了.
# 所以我们需要 \b 来实现
print(ans)

# 注意这里: 此时有了 \b 那么对应的, cati!  caterpillar  就不能匹配上了, 因为这里有单词原子, 单词原子: [a-zA-Z0-9]
s = "The cat sat on the caterpillar. cati!"

ans = re.findall(r"cat\b", s)

print(ans)

# 同样的, 出现了 lcat 那么对应的我们应该在 reg 的开头也匹配上 \b
s = "lcat The cat sat on the caterpillar. cati!"

ans = re.findall(r"\bcat\b", s)

print(ans)
````

#### 11.2 取消特殊功能符号以普通化

````python
s = "https://wwwwxxx.com/,https://www.baidu.com/,yuan,rain,http://www.jd.com/,https://www.taobao.com/"

# 注意: 这个匹配字符串有 bug 因为其中的 www. 中的 . 也是正则表达式中的通配符.  导致wwwwxxx也被匹配上了, 而这并不是我们想要的.
ans = re.findall("https??://www.[a-z]*?.com", s)
# 所以此时我们将通配符 . 转义为普通的 .
ret = re.findall(r"https??://www\.[a-z]*?\.com", s)

print(ans)
print(ret)
````

## 12 () 分组与优先提取

````python
s = "apple banana peach orange aaa appleappleapple appleapple"

# 这里主要实现的分组的功能.  这里的(apple) 将 apple 作为一个整体被解析, 而不是将最后的 e 原子字符被 {} 重复选择符解析
ret = re.findall("(?:apple){2,3}", s)
print(ret)
````

````python
text = """
Visit us at user@qq.com for more info.
Contact support at support@qq.com.
Also, check out admin@my163.com and info@163.com
"""
# 这里展示了优先提取的功能.  我们只是想要对应的提取, 并不想要多余的东西.
ret = re.findall(r"\b([\w.-]+)@\w*163\.com\b", text)

print(ret)
````

### 13 | 或 与字符集 [] 类似, 但是字符集是将一个原子字符进行选择, | 是将一个大整体进行选择.

````python
text = """
Visit us at example.com for more information.
You can also check out mysite.cn and partner-site.com.
Don't forget about blog-example.com or support@service.cn.
"""

# ans = re.findall(r"\b[\w.-]+\.com", text)
# ans = re.findall(r"\b[\w.-]+\.cn", text)

# 这里设置两组匹配数据, 用 | 链接 (但是这样设置太繁琐了)
# ans = re.findall(r"\b[\w.-]+\.cn|\b[\w.-]+\.com", text)

ans = re.findall(r"\b[\w.-]+\.(com|cn)", text)

# 所以这里我们需要取消优先提取的功能, 在括号最前加入: ?:
ans1 = re.findall(r"\b[\w.-]+\.(?:com|cn)", text)

# 这里我们只想要前面的域名, 并不想要最后的.com .cn 那么我们可以直接使用()的优先提取功能, 直接实现提取.
ans2 = re.findall(r"\b([\w.-]+)\.(?:com|cn)", text)
print(ans)
print(ans1)
print(ans2)
````

## 正则表达式模块 re 中的常用函数

findall()方法 和 search()方法

findall()方法是找到**所有的匹配**. def findall(pattern, string, flags=0)
search()方法是找到**第一个匹配**.  def search(pattern, string, flags=0)
