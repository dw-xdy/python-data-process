***Using Lazy loading For 1+ Billion Data***

```python
import polars as pl

# Lazy loading the Parquet file
df_lazy = pl.scan_parquet(r"F:\E\python\python数据分析的数据\2021\yellow\*").select(
    ["trip_distance", "store_and_fwd_flag", "total_amount"]
)
```

```python
# Collect the result to materialize it as a DataFrame
df_lazy = df_lazy.collect(engine="streaming")
```

```python
# Display the result
df_lazy
```
